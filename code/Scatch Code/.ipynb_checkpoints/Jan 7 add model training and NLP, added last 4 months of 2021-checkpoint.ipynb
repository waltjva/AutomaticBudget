```python
import os
import pandas as pd
os.chdir('..')
from auto_budget import bank_functions_NWFCU as nw

# historic data back to March 2020 (COVID-19 pandemic) with categories
data = pd.read_excel('../data/CleanedStatements/fully_catized_encoded.xlsx')

# onboarding new purchases from NW Federal Credit Union checking account
filename = '../data/RawStatements/ExportedTransactions-8.csv'
new_data = nw.clean_bank_statement_file(filename)
new_data.Cat_Self_Pred = nw.categorize(new_data)
coded_data = nw.encode_data(new_data)
coded_data.Category_Pred = nw.decision_tree_new_statements(data, coded_data)
ref = nw.ref_dictionaries(data)
coded_data.Cat_Pred_Words = coded_data.Category_Pred.apply(lambda x: ref[1][x])
coded_data.Cat_Self_Pred = nw.self_categorization(coded_data, ref)
coded_data
```

```python
coded_data.to_csv('categorized_for_now.csv',index=False)
```

```python
# in the new data but not in the old
lis = []
for x in coded_data.columns:
    if x not in data.columns:
        lis.append(x)
lis
```

```python
coded_data.Self_Pred_Nums = coded_data.Cat_Self_Pred.apply(lambda x: ref[0][x])
```

```python
to_be_joined = coded_data.copy()
to_be_joined['Category'] = to_be_joined['Cat_Self_Pred'].copy()
to_be_joined['Category_Cat'] = to_be_joined['Self_Pred_Nums'].copy()
to_be_joined.drop(lis,axis=1,inplace=True)
to_be_joined
```

```python
cols = list(data.columns)
to_be_joined = to_be_joined[cols].copy()
to_be_joined.columns == data.columns
```
