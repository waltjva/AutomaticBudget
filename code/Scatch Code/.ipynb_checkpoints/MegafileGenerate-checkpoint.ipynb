```python
import pandas as pd
import numpy as np
```

```python
mar_df = pd.read_csv('MarchChecking.csv')
apr_df = pd.read_csv('AprilChecking.csv')
may_df = pd.read_csv('MayChecking.csv')
jun_df = pd.read_csv('JuneChecking.csv')
```

```python
mar_df.columns = ['Transaction Number','Date','Description','Memo','Amount Debit','Amount Credit','Balance',
                 'Check Number','Fees']
apr_df.columns = ['Transaction Number','Date','Description','Memo','Amount Debit','Amount Credit','Balance',
                 'Check Number','Fees']
may_df.columns = ['Transaction Number','Date','Description','Memo','Amount Debit','Amount Credit','Balance',
                 'Check Number','Fees']
jun_df.columns = ['Transaction Number','Date','Description','Memo','Amount Debit','Amount Credit','Balance',
                 'Check Number','Fees']
```

```python
mar_df.drop([0,1,2], inplace=True)
apr_df.drop([0,1,2], inplace=True)
may_df.drop([0,1,2], inplace=True)
jun_df.drop([0,1,2], inplace=True)
```

```python
mar_df.reset_index(drop=True,inplace=True)
apr_df.reset_index(drop=True,inplace=True)
may_df.reset_index(drop=True,inplace=True)
jun_df.reset_index(drop=True,inplace=True)
```

```python
def debit_credit(debit,credit):
    return debit+credit
```

```python
mar_df['Amount Debit'].fillna(0,inplace=True)
mar_df['Amount Debit'] = mar_df['Amount Debit'].astype(float)
mar_df['Amount Credit'].fillna(0,inplace=True)
mar_df['Amount Credit'] = mar_df['Amount Credit'].astype(float)

apr_df['Amount Debit'].fillna(0,inplace=True)
apr_df['Amount Debit'] = mar_df['Amount Debit'].astype(float)
apr_df['Amount Credit'].fillna(0,inplace=True)
apr_df['Amount Credit'] = mar_df['Amount Credit'].astype(float)

may_df['Amount Debit'].fillna(0,inplace=True)
may_df['Amount Debit'] = mar_df['Amount Debit'].astype(float)
may_df['Amount Credit'].fillna(0,inplace=True)
may_df['Amount Credit'] = mar_df['Amount Credit'].astype(float)

jun_df['Amount Debit'].fillna(0,inplace=True)
jun_df['Amount Debit'] = mar_df['Amount Debit'].astype(float)
jun_df['Amount Credit'].fillna(0,inplace=True)
jun_df['Amount Credit'] = mar_df['Amount Credit'].astype(float)
```

```python
mar_df['Amount'] = debit_credit(mar_df['Amount Debit'],mar_df['Amount Credit'])
apr_df['Amount'] = debit_credit(apr_df['Amount Debit'],apr_df['Amount Credit'])
may_df['Amount'] = debit_credit(may_df['Amount Debit'],may_df['Amount Credit'])
jun_df['Amount'] = debit_credit(jun_df['Amount Debit'],jun_df['Amount Credit'])
```

```python
mar_df.drop(['Transaction Number','Amount Debit','Amount Credit','Balance','Check Number','Fees'],axis=1,inplace=True)
apr_df.drop(['Transaction Number','Amount Debit','Amount Credit','Balance','Check Number','Fees'],axis=1,inplace=True)
may_df.drop(['Transaction Number','Amount Debit','Amount Credit','Balance','Check Number','Fees'],axis=1,inplace=True)
jun_df.drop(['Transaction Number','Amount Debit','Amount Credit','Balance','Check Number','Fees'],axis=1,inplace=True)
```

```python
mar_df['Category'] = 0
apr_df['Category'] = 0
may_df['Category'] = 0
jun_df['Category'] = 0
```

```python
def statement_parse(description,category,amount):
    for x in range(len(description)):
        if(description[x] == 'FEDLOANSERVICING/STDNT' or description[x] == 'FEDLOANSERVICING/STDNT LOANWALTE'):
            category[x] = 1
        #Car payment
        #Unnecessary if
        #USAA Insurance
        elif(description[x] == 'USAA.COM PAY EXT/P&CWALTER COLEM' or description[x] == 'USAA.COM PAY EXT/P&CWALTER'):
            category[x] = 3
        #Rent
        elif(description[x] == 'VENMO/PAYMENTWALTER COLEMAN' and amount[x] == -668.75):
            category[x] = 4
        #Phone
        elif(description[x] == 'TOM COLEMAN/BILL PAYMTCOLEMAN'):
            category[x] = 5
        #Gas
        elif(description[x] == 'POS PURCHASE Non-PINCOLUMBIA'):
            category[x] = 6
        #Utilities
        elif(description[x] == 'WASHINGTON GAS/PAYMENTEvan' or description[x] == 'DOM ENERGYVA/UTILITYWALTER' or
            description[x] == 'VERIZON/PAYMENTRECWALTER COLEMAN' or description[x] == 'DOM ENERGYVA/UTILITYWALTER *COLE'):
            category[x] = 7
        #Tolls/Metro/Uber
        elif(description[x] == 'UberNotReally'):
            category[x] = 8
        #Dining Out
        elif(description[x] == 'POS PURCHASE Non-PINUBER EATS' or description[x] == 'POS PURCHASE Non-PINSQ *SOL'
            or description[x] == 'POS PURCHASE Non-PINUberAT_EATS'):
            category[x] = 9
        #Groceries
        elif(description[x] == 'POS PURCHASE with PINGIANT 0774' or description[x] == 'POS PURCHASE with PINHARRIS'):
            category[x] = 10
        #Entertainment
        elif(description[x] == 'POS PURCHASE Non-PINPP*GOOGLE A'):
            category[x] = 11
        #Tithe 
        elif(description[x] == 'Pillar DC/Pillar DCPILLAR' or description[x] == 'POS PURCHASE Non-PINPILLAR DC'):
            category[x] = 12
        #Medical
        elif(description[x] == 'POS PURCHASE Non-PINVTG*Emmaus' or description[x] == 'POS PURCHASE Non-PINLCA*LABCORP'):
            category[x] = 13
        #401k
        #Unnecessary as it is already calculated in
        #Maintenance Service
        elif(description[x] == 'Maintenance Service Fee'):
            category[x] = 15
        #Jiu Jitsu & Gym
        elif(description[x] == 'POS PURCHASE Non-PINDISTRICT' or description[x] == 'POS PURCHASE Non-PINXSPORT'):
            category[x] = 16
        #Venmo extra
        elif((description[x] == 'VENMO/CASHOUTWALTER COLEMAN' or description[x] == 'Person-to-Person') and amount[x] > 0):
            category[x] = 18
        #Accenture payceck
        elif(description[x] == 'Accenture Federa/PAYRLL' and amount[x] > 0):
            category[x] = 19
        #Misc spending
        elif(amount[x] < 0 and description[x] != 'INTERNET TRANSFER TO: DDXXXX5941'):
            category[x] = 17
        #Misc income
        elif(amount[x] > 0 and description[x] != 'INTERNET TRANSFER FROM:'):
            category[x] = 20
```

```python
statement_parse(mar_df['Description'],mar_df['Category'],mar_df['Amount'])
statement_parse(apr_df['Description'],apr_df['Category'],apr_df['Amount'])
statement_parse(may_df['Description'],may_df['Category'],may_df['Amount'])
statement_parse(jun_df['Description'],jun_df['Category'],jun_df['Amount'])
```
