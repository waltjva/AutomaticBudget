```python
def debit_credit(debit,credit):
    return debit+credit
```

```python
category = 0
```

```python
def statement_parse(description,category,amount):
    for x in range(len(description)):
        print("iteration: ", x)
        #Student Loans
        if(description[x] == 'FEDLOANSERVICING/STDNT' or description[x] == 'FEDLOANSERVICING/STDNT LOANWALTE'):
            category[x] = 1
        #Car payment
        #Unnecessary if
        #USAA Insurance
        elif(description[x] == 'USAA.COM PAY EXT/P&CWALTER COLEM' or description[x] == 'USAA.COM PAY EXT/P&CWALTER' or
            description[x] == 'USAA P&C/PAYMENTWALTER COLEMAN'):
            category[x] = 3
        #Rent
        elif(description[x] == 'VENMO/PAYMENTWALTER COLEMAN' and amount[x] == -668.75):
            category[x] = 4
        #Phone
        elif(description[x] == 'TOM COLEMAN/BILL PAYMTCOLEMAN'):
            category[x] = 5
        #Gas
        elif(description[x] == 'POS PURCHASE Non-PINCOLUMBIA' or description[x] == 'POS PURCHASE Non-PINCOSTCO GAS' or 
             description[x] == 'POS PURCHASE Non-PINSHELL'):
            category[x] = 6
        #Utilities
        elif(description[x] == 'WASHINGTON GAS/PAYMENTEvan' or description[x] == 'DOM ENERGYVA/UTILITYWALTER' or
            description[x] == 'VERIZON/PAYMENTRECWALTER COLEMAN' or description[x] == 'DOM ENERGYVA/UTILITYWALTER *COLE'):
            category[x] = 7
        #Tolls/Metro/Uber
        elif(description[x] == 'POS PURCHASE Non-PINARLINGTON'):
            category[x] = 8
        #Dining Out
        elif(description[x] == 'POS PURCHASE Non-PINUBER EATS' or description[x] == 'POS PURCHASE Non-PINSQ *SOL' or 
             description[x] == 'POS PURCHASE Non-PINUberAT_EATS'):
            category[x] = 9
        #Groceries
        elif(description[x] == 'POS PURCHASE with PINGIANT 0774' or description[x] == 'POS PURCHASE with PINHARRIS' or 
            description[x] == 'POS PURCHASE with PINALDI 71180'):
            category[x] = 10
        #Entertainment
        elif(description[x] == 'POS PURCHASE Non-PINPP*GOOGLE A'):
            category[x] = 11
        #Tithe 
        elif(description[x] == 'Pillar DC/Pillar DCPILLAR' or description[x] == 'POS PURCHASE Non-PINPILLAR DC' or 
             description[x] == 'Pillar DC/Pillar DCPILLAR CHURCH'):
            category[x] = 12
        #Medical
        elif(description[x] == 'POS PURCHASE Non-PINVTG*Emmaus' or description[x] == 'POS PURCHASE Non-PINLCA*LABCORP'
            or description[x] == 'POS PURCHASE Non-PINADVANCED'):
            category[x] = 13
        #401k
        #Unnecessary as it is already calculated in
        #Maintenance Service
        elif(description[x] == 'Maintenance Service Fee'):
            category[x] = 15
        #Jiu Jitsu & Gym
        elif(description[x] == 'POS PURCHASE Non-PINDISTRICT' or description[x] == 'POS PURCHASE Non-PINXSPORT'):
            category[x] = 16
        #Venmo extra
        elif((description[x] == 'VENMO/CASHOUTWALTER COLEMAN' or description[x] == 'Person-to-Person') and amount[x] > 0):
            category[x] = 18
        #Accenture payceck
        elif(description[x] == 'Accenture Federa/PAYRLL' and amount[x] > 0):
            category[x] = 19
        #Misc spending
        elif(amount[x] < 0 and description[x] != 'INTERNET TRANSFER TO: DDXXXX5941'):
            category[x] = 17
        #Misc income
        elif(amount[x] > 0 and description[x] != 'INTERNET TRANSFER FROM:'):
            category[x] = 20
    return category
```

```python
mar_df['cat'] = statement_parse(mar_df['Description'].copy(),mar_df['Category'].copy(),mar_df['Amount'].copy())
```
