```python
import pandas as pd
oct_df = pd.read_csv('OctoberChecking.csv')
oct_df.columns = oct_df.loc[2]
oct_df.drop([0,1,2],inplace=True)
oct_df['Amount Debit'].fillna(0,inplace=True)
oct_df['Amount Credit'].fillna(0,inplace=True)
oct_df['Memo'].fillna('Default',inplace=True)
oct_df['Transaction Number'] = oct_df['Transaction Number'].apply(lambda x: x.split('**')[1])
oct_df['Memo'] = oct_df['Memo'].apply(lambda x: x.split('*****')[0])
oct_df['Content'] = oct_df['Transaction Number'] + ' / ' + oct_df['Memo']
oct_df['Amount'] = oct_df['Amount Debit'].astype(float) + oct_df['Amount Credit'].astype(float)
oct_df['Category'] = 'Default'
oct_df.drop(oct_df[oct_df['Content'].str.contains("INTERNET TRANSFER",na=False)].index,inplace=True)
oct_df.reset_index(inplace=True)
oct_df.drop(['index','Transaction Number','Date','Description','Memo','Amount Debit','Amount Credit',
             'Balance','Check Number','Fees  '],axis=1,inplace=True)
pd.set_option('display.max_rows',None)
pd.set_option('display.max_rows',None)
pd.set_option('display.width', 1000)
```
